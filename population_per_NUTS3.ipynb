```python
import pandas as pd
```

```python
# Read data
data = pd.read_csv('/Users/maijahavusela/Desktop/gradu/data/population_NUTS3.csv',sep=',')
data.head()
```

```python
# Choose wanted columns
data = data[['geo','TIME_PERIOD','OBS_VALUE']]

# Clean column names
data['NUTS_CODE'] = data['geo'].str.split(':').str[0]
data['NUTS_NAME'] = data['geo'].str.split(':').str[-1]

# Renaming columns
data.rename(columns={"OBS_VALUE": "POP"}, inplace=True)
data = data[['NUTS_CODE','NUTS_NAME','POP','TIME_PERIOD']]
data
```

## Aggregating to area (square kilometers)

```python
# Read data
data2 = pd.read_csv('/Users/maijahavusela/Desktop/gradu/data/area_NUTS3.csv',sep=',')

# Choose wanted columns
data2 = data2[['geo','TIME_PERIOD','OBS_VALUE']]

# Renaming columns
data2.rename(columns={"OBS_VALUE": "AREA", "geo":"NUTS_NAME"}, inplace=True)
data2.head()
```

```python
# Normalizing names
data2["NUTS_NAME"] = data2["NUTS_NAME"].str.strip().str.lower()
data["NUTS_NAME"] = data["NUTS_NAME"].str.strip().str.lower()

# Aggregating the population dataframe with the area dataframe
df_merged = data2.merge(data, left_on="NUTS_NAME", right_on="NUTS_NAME", how="left")
df_merged
```
